The purpose of this Jupyter notebook is to gather descriptive data of the datasets before applying models. This includes bar graphs of mean test scores by counties, school level, and school type (charter vs. traditional), along with possibly other graphical representations of the data. 

```python
# imports for the notebook
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

A bar graph that conveys average percent passing between charter and traditional schools

```python
# reads in CSV and obtains school type and percent passing columns
df_combined = pd.read_csv('combined_data/combined.csv')
df_combined_2022 = pd.read_csv('yearly_data/combined_2022.csv')
df_charter_traditional = df_combined[['Charter School Dummy', 'Average Percent Passing']]

# calculates mean and std for the two groups
groups = df_charter_traditional.groupby('Charter School Dummy')
group_means = groups.mean()
group_stds = groups.std().values.reshape(-1)

# generates bar graph
labels = ['Traditional School', 'Charter School']
plt.bar(group_means.index, group_means['Average Percent Passing'], yerr=group_stds, capsize=5)
plt.title('Average Percent Passing by School Type')
plt.xlabel('School Type')
plt.ylabel('Average Percent Passing')
plt.xticks(group_means.index, labels)
plt.show()
```

A bar graph that conveys average percent passing by school type

```python
# obtains school level and percent passing columns
df_school_type = df_combined_2022[['School Level', 'Average Percent Passing']]

# calculates mean and std for the two groups
groups = df_school_type.groupby('School Level')
group_means = groups.mean()
group_stds = groups.std().values.reshape(-1)

# Create a bar plot with error bars
labels = ['Elementary', 'Middle', 'High', 'Other']
plt.bar(group_means.index, group_means['Average Percent Passing'], yerr=group_stds, capsize=5)
plt.title('Average Percent Passing in 2022 by School Level')
plt.xlabel('School Level')
plt.ylabel('Average Percent Passing')
plt.xticks(group_means.index, labels)
plt.show()
```
